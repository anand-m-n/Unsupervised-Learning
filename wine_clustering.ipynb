#### Scaling

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
data_sc = scalar.fit_transform(data)
```

```python
data_sc
```

```python
X = pd.DataFrame(data = data_sc, columns=data.columns)
```

```python
X
```

```python
X.describe()
```

#### K-means clustering

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
wcss
```

```python
plt.plot(range(1,11), wcss)
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
```

From the elbow method, number of clusters = 3

```python
kmeans = KMeans(n_clusters=3, random_state=42, init='k-means++')
y_means = kmeans.fit_predict(X)
```

```python
y_means
```

#### Hierarchical clustering

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
```

From the dendrogram, the number of clusters = 3

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
y_hc = hc.fit_predict(X)
y_hc
```
